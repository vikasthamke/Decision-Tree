## Importing Libraries

```python
import pandas as pd
```

## Reading our CSV File

```python
import requests
url = 'https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/Exercise/titanic.csv'
res = requests.get(url,allow_redirects = True)
with open('titanic.csv','wb') as file:
    file.write(res.content)
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df['Age'].unique()
```

```python
#df.dropna(inplace=True)
df.Age.fillna(df.Age.mean(),inplace=True)
```

```python
df['Age'].unique()
```

```python
inputs = df.drop('Survived',axis=1)
inputs.head()
```

```python
target = df['Survived']
target
```

## Encodeing 'Sex' feature

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder()
```

```python
inputs['sex_n'] = le_sex.fit_transform(inputs['Sex'])
inputs.drop(['Sex','Name','Ticket','Cabin','Embarked','PassengerId'],axis=1,inplace=True)
inputs.head()
```

## Splitting Data into Training and Testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(inputs,target,test_size=0.2)

print("Shape of x_train :", x_train.shape)
print("Shape of y_train :", y_train.shape)
print("Shape of x_test :", x_test.shape)
print("Shape of y_test :", y_test.shape)
```

## Train Our Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
```

```python
tree.fit(x_train,y_train)
```

## Accuracy Of Our Model

```python
tree.score(x_test,y_test)
```
